```python
target = 'species'
y_train = df_train[target]
X_train = df_train.drop(columns=[target, 'id'])
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X_train = StandardScaler().fit(X_train).transform(X_train)
```

```python
y_train = LabelEncoder().fit(y_train).transform(y_train)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train
```

```python
y_train.shape
```

```python
from keras.utils.np_utils import to_categorical
y_train = to_categorical(y_train)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=.3)
```

```python
input_shape = X_train.shape[-1]
output_shape = y_train.shape[-1]
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,10
```

```python
# 1) logging files 
# create 2 dir -- one for tensorboard results and one for early stopping
logdir = os.path.join('logs', 'NoEarlyStopping')


# 2) Callbacks 
# instantiate a tensorboard callback object

# A Callback is a function that is called after some event during training, or evaluation (for example, before or after each training epoch)
# Tensorboard Logs Callback for saving our data
tensorboard_callback = TensorBoard(logdir, histogram_freq=1)

# 3) Build the model 
def create_model():
    
# instantiate Sequential class
    model = Sequential()

# flatten images (this also serves as the input_dim)
    model.add(Dense(1024,input_dim=192))

# hidden layer 1
    model.add(Dense(512))   
# activation function 1
    model.add(ReLU(negative_slope=.01))


# hidden layer 1
    model.add(Dense(256))   
# activation function 1
    model.add(ReLU(negative_slope=.01))

# hidden layer 1
    model.add(Dense(128))   
# activation function 1
    model.add(ReLU(negative_slope=.01))

# output layer 
    model.add(Dense(99, activation='softmax'))

# compile model 
    model.compile(optimizer='adam', 
                  loss='categorical_crossentropy', 
                  metrics=['accuracy'])

# fit model 
    return model

# build and compile the model

model = create_model()

# fit the model
model.fit(X_train, 
          y_train, 
          epochs=20, 
          validation_data=(X_val, y_val), 
          verbose=1, 
          callbacks=[tensorboard_callback])
```

```python
%tensorboard --logdir logs
```

```python
# 1) logging files 
# create 2 dir -- one for tensorboard results and one for early stopping
logdir = os.path.join('logs', 'EarlyStopping')


# 2) Callbacks 
# instantiate a tensorboard callback object

# A Callback is a function that is called after some event during training, or evaluation (for example, before or after each training epoch)
# Tensorboard Logs Callback for saving our data
tensorboard_callback = TensorBoard(logdir, histogram_freq=1)

# Early Stopping Callback
# ALWAYS do early stopping on the validation metric
stop = EarlyStopping(monitor='val_loss', min_delta=0.01, patience=3)

# build and compile the model

model = create_model()

# fit the model
model.fit(X_train, 
          y_train, 
          epochs=20, 
          validation_data=(X_val, y_val), 
          verbose=1, 
          callbacks=[tensorboard_callback, stop])
```
